```python
def get_path(imgid):
    return str(img_dir) + '/' + df_final[df_final.image_id == imgid].file_name[0]
```

```python
get_path(test_id)
```

```python
def get_first_bbox(imgid):
    return df_final[df_final.image_id == imgid].bbox[0]
```

```python
get_first_bbox(test_id)
```

```python
import matplotlib.patches as patches
import skimage.io as io
```

```python
def plt_img_bbox(imgid):
    img = get_path(imgid)
    bbox = get_first_bbox(imgid)
    rect = patches.Rectangle((bbox[0], bbox[1]), bbox[2], bbox[3], color='r')
    f, ax = plt.subplots(figsize=(10,10))
    ax.imshow(io.imread(img))
    ax.add_artist(rect)
    plt.show()
```

```python
for img in list(df_final['image_id'].values)[:5]:
    plt_img_bbox(img)
```

```python
I = plt.imread(get_path(test_id))
plt.figure(figsize=(10,10))
plt.imshow(I)
```

```python
dls = dblock.dataloaders(df_new, bs=16)
```

```python
path_pascal = untar_data(URLs.PASCAL_2007)
```

```python
path_pascal
```

```python
annots = json.load(open(path_pascal/'train.json'))
```

```python
annots.keys()
```

```python
train_images, train_lbl_bbox = get_annotations(path_pascal/'train.json')
```

```python
train_images[0], train_lbl_bbox[0]
```

```python
images, lbl_bbox = train_images,train_lbl_bbox
img2bbox = dict(zip(images, lbl_bbox))
```

```python
getters = [lambda o: path_pascal/'train'/o, lambda o: img2bbox[o][0], lambda o: img2bbox[o][1]]
```

```python
item_tfms = [Resize(128, method='pad'),]
batch_tfms = [Rotate(), Flip(), Dihedral(), Normalize.from_stats(*imagenet_stats)]
```

```python
def get_train_imgs(noop):  return images
```

```python
dblock = DataBlock(blocks=(ImageBlock, BBoxBlock, BBoxLblBlock),
                 get_items=get_train_imgs,
                   getters=getters,
                 splitter=RandomSplitter(),
                 #get_y=[lambda o: img2bbox[o.name][0], lambda o: img2bbox[o.name][1]], 
                 item_tfms=Resize(256),
                 batch_tfms=aug_transforms(),
                 n_inp=1)
```

```python
dls = dblock.dataloaders(path_pascal/'train')
#dls.show_batch()
get_c(dls)
```

```python
dls.one_batch()
```

```python
dls.show_batch()
```

```python
encoder = create_body(xresnet34, pretrained=True)
```

```python
arch = RetinaNet(encoder, 20, final_bias=-4)
```

```python
create_head(124, 4)
```

```python
ratios = [1/2,1,2]
scales = [1,2**(-1/3), 2**(-2/3)]
```

```python
crit = RetinaNetFocalLoss(scales=scales, ratios=ratios)
```

```python
def _retinanet_split(m): 
    return L(m.encoder,nn.Sequential(m.c5top6, m.p6top7, m.merges, m.smoothers, m.classifier, m.box_regressor)).map(params)
```

```python
learn = Learner(dls, arch, loss_func=crit, splitter=_retinanet_split)
```

```python
learn.freeze()
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(10, slice(1e-4, 1e-3))
```
